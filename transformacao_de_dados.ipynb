```python
import pandas as pd
import pandera  as pa  # biblioteca para validação de extração de dados
```

```python
valores_ausentes = ['**', '###!','####','****','*****','NULL']
df = pd.read_csv("ocorrencia_2010_2020.csv", sep=";", parse_dates=["ocorrencia_dia"], dayfirst=True, na_values=valores_ausentes)
df.head()
```

```python
# Validação de dados das Colunas- Schema
schema = pa.DataFrameSchema(
    columns = {
        "codigo": pa.Column(pa.Int, required=False),
        "codigo_ocorrencia":pa.Column(pa.Int),
        "codigo_ocorrencia2":pa.Column(pa.Int),
        "ocorrencia_classificacao":pa.Column(pa.String),
        "ocorrencia_uf":pa.Column(pa.String, nullable=True),#pa.Check.str_length(2,2),),
        "ocorrencia_aerodromo":pa.Column(pa.String, nullable=True),
        "ocorrencia_dia":pa.Column(pa.DateTime),
        "ocorrencia_hora":pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True),
         "total_recomendacoes":pa.Column(pa.Int)
    }
)
```

```python
schema.validate(df)
```

```python
#  ocorrências com (classificação == INCIDENTE GRAVE  ou   classificação == INCIDENTE) e o  estado de SP
filtro1 = (df.ocorrencia_classificacao == 'INCIDENTE GRAVE') | (df.ocorrencia_classificacao == 'INCIDENTE')
filtro2 = df.ocorrencia_uf == 'SP'
df.loc[filtro1 &  filtro2]
```

```python
# ocorrências por Ano
filtro_ano = df.ocorrencia_dia.dt.year == 2020
df.loc[filtro_ano]
```

```python
# ocorrências por Ano e mês
filtro_ano = df.ocorrencia_dia.dt.year == 2020
filtro_mes = df.ocorrencia_dia.dt.month == 12
df.loc[filtro_ano & filtro_mes]
```

```python
# ocorrências por Ano, mês e dia
filtro_ano = df.ocorrencia_dia.dt.year == 2020
filtro_mes = df.ocorrencia_dia.dt.month == 12
filtro_dia = df.ocorrencia_dia.dt.day == 21
df.loc[filtro_ano & filtro_mes & filtro_dia]
```

```python
# ocorrências por Ano, mês e intervalos de  dia
filtro_ano = df.ocorrencia_dia.dt.year == 2020
filtro_mes = df.ocorrencia_dia.dt.month == 12
filtro_dia_inicio = df.ocorrencia_dia.dt.day > 9 
filtro_dia_fim = df.ocorrencia_dia.dt.day < 20 
df.loc[filtro_ano & filtro_mes & filtro_dia_inicio & filtro_dia_fim]
```

```python
# coluna ocorrencia_dia_hora
df['ocorrencia_dia_hora'] = pd.to_datetime(df.ocorrencia_dia.astype(str) + ' ' + df.ocorrencia_hora)
```
